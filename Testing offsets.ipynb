```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from structure import build_price
```

```python
def plot_cost_function(price, num_breaks):
    offsets = build_price(price, num_breaks)
    print(offsets)
    for i in range(price.shape[0]):
        print('Iteration: ', i)
        slope = price[i, :num_breaks]
        breakpoints = price[i, num_breaks:]
        offset = offsets[i]
        print('The offsets are:', offset)
        m = breakpoints.min()
        m = m * 2 if m != 0 else -3
        M = breakpoints.max()
        M = M * 2 if M != 0 else 3
        X = np.arange(m, M, 0.1).reshape(1, -1)
        f = np.dot(slope.reshape(-1, 1), X) + offset.reshape(-1, 1)
        fig, ax = plt.subplots()
        ax.plot(X.reshape(-1, 1), f.max(axis=0))
        ax.axhline(0, c='k')
        ax.axvline(0, c='k')
        for b in breakpoints:
            ax.axvline(b, c='r')
```

```python
price = np.array([[1, 3, 6, 10, -3, 4, 5],
                [1,3, 6, 10, -3, 0, 5]])
plot_cost_function(price, 4)
```

```python
price = np.array([[1, 3, 0]])
plot_cost_function(price, 2)
```

```python
price = np.array([[1, 1.5, 3.7, 4, -0.6, 0, 0]])
plot_cost_function(price, 4)
```
